```python
df_filtered = df.drop(columns=['OBJECTID', 'WBANNO', 'LST_YRMO', 'LST_YRMO_Converted'])

correlation_matrix = df_filtered.corr()

plt.figure(figsize=(24, 20))
sns.heatmap(correlation_matrix**2, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={'label': 'RÂ² Value'})

plt.title('RÂ² Values Heatmap', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Get correlations with the dependent variable T_MONTHLY
target_corr = correlation_matrix['T_MONTHLY'].abs()

# Select features strongly correlated with T_MONTHLY (rÂ² > 0.5)
strong_features = target_corr[target_corr**2 > 0.5].index.tolist()
strong_features.remove('T_MONTHLY')  # Remove the dependent variable itself

print("Strong features (rÂ² > 0.5 with T_MONTHLY):", strong_features)

# Now check pairwise correlations among the selected strong features
if strong_features:
    selected_corr_matrix = correlation_matrix.loc[strong_features, strong_features]

    # Filter out features with high multicollinearity (rÂ² < 0.5 between features)
    low_multicollinearity_features = []

    for feature in strong_features:
        # Check if this feature has low multicollinearity with all features already selected
        if selected_corr_matrix[feature].apply(lambda x: x**2 < 0.5).all():
            low_multicollinearity_features.append(feature)

    print("Low multicollinearity features (rÂ² < 0.5 between features):", low_multicollinearity_features)
else:
    print("No features have strong correlation with T_MONTHLY.")

# Add the dependent variable back for use in the model
final_features = low_multicollinearity_features + ['T_MONTHLY']

# Display the selected features
print("Selected features for the regression model:", final_features)
```
